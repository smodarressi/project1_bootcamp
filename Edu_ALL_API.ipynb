```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from pprint import pprint
from api_keys import datagov_api_key
```

```python
all_fields_df = pd.read_csv('all_fields.csv')
all_fields_df

all_fields_str = ""

for row in all_fields_df.index:
    if row == 0:
        all_fields_str += f"latest.{all_fields_df.loc[row,\
        'dev-category']}.{all_fields_df.loc[row, 'developer-friendly name']}"
    else:
        all_fields_str += f",latest.{all_fields_df.loc[row,\
        'dev-category']}.{all_fields_df.loc[row, 'developer-friendly name']}"

all_fields_str
```

```python
#latest average net cost is from 2016
base_url = 'https://api.data.gov/ed/collegescorecard/v1/schools.json'

params = {'api_key': datagov_api_key,
          'school.institutional_characteristics.level': '1',
          'school.online_only': '0',
          '_fields': f'id,school.name,location.lat,location.lon,{all_fields_str}',
          '_per_page': '100',
          '_page': 0
          }

response = requests.get(base_url, params=params)
response_json = response.json()
```

```python
print(response.url)
```

```python
pprint(response_json)
```

```python
response_df = pd.DataFrame(response_json['results'])

response_df.head()
```

```python
num_pages = round(response_json['metadata']['total'] / response_json['metadata']['per_page'])
num_pages
```

```python
for i in range(1,num_pages + 1):
    params = {'api_key': datagov_api_key,
          'school.institutional_characteristics.level': '1',
          'school.online_only': '0',
          '_fields': f'id,school.name,location.lat,location.lon,{all_fields_str}',
          '_per_page': '100',
          '_page': i
          }
    response = requests.get(base_url, params=params)
    response_json = response.json()
    next_response_df = pd.DataFrame(response_json['results'])
    
    response_df = response_df.append(next_response_df)
```
